```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parametry modelu Lotka-Volterra
alpha = 0.1   # Rychlost růstu kořisti
beta = 0.02   # Míra lovu kořisti dravci
delta = 0.01  # Míra růstu dravců díky kořisti
gamma = 0.1   # Úmrtnost dravců

# Definice soustavy rovnic Lotka-Volterra
def lotka_volterra(t, z):
    x, y = z  # x = kořist, y = dravci
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

# Počáteční podmínky: populace kořisti a dravců
x0 = 40  # Počáteční populace kořisti
y0 = 9   # Počáteční populace dravců
z0 = [x0, y0]

# Časový interval simulace
t_span = (0, 200)  # časový interval od 0 do 200
t_eval = np.linspace(0, 200, 1000)  # hodnoty času pro vykreslení

# Numerické řešení soustavy rovnic
sol = solve_ivp(lotka_volterra, t_span, z0, t_eval=t_eval, method='RK45')

# Výsledné hodnoty pro kořist a dravce
x = sol.y[0]  # populace kořisti
y = sol.y[1]  # populace dravců

# Vizualizace výsledků
plt.figure(figsize=(10, 6))
plt.plot(sol.t, x, label='Kořist (x)', color='blue')
plt.plot(sol.t, y, label='Dravci (y)', color='red')
plt.title('Model Lotka-Volterra: Populace dravců a kořisti v čase')
plt.xlabel('Čas')
plt.ylabel('Populace')
plt.legend()
plt.grid(True)
plt.show()

# Fázový diagram (kořist vs. dravci)
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green')
plt.title('Fázový diagram: Kořist vs. Dravci')
plt.xlabel('Populace kořisti (x)')
plt.ylabel('Populace dravců (y)')
plt.grid(True)
plt.show()
```
