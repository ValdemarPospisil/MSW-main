```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parametry modelu Lotka-Volterra
alpha = 0.1   # Rychlost rstu koisti
beta = 0.02   # Mra lovu koisti dravci
delta = 0.01  # Mra rstu dravc dky koisti
gamma = 0.1   # mrtnost dravc

# Definice soustavy rovnic Lotka-Volterra
def lotka_volterra(t, z):
    x, y = z  # x = koist, y = dravci
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

# Poten podmnky: populace koisti a dravc
x0 = 40  # Poten populace koisti
y0 = 9   # Poten populace dravc
z0 = [x0, y0]

# asov interval simulace
t_span = (0, 200)  # asov interval od 0 do 200
t_eval = np.linspace(0, 200, 1000)  # hodnoty asu pro vykreslen

# Numerick een soustavy rovnic
sol = solve_ivp(lotka_volterra, t_span, z0, t_eval=t_eval, method='RK45')

# Vsledn hodnoty pro koist a dravce
x = sol.y[0]  # populace koisti
y = sol.y[1]  # populace dravc

# Vizualizace vsledk
plt.figure(figsize=(10, 6))
plt.plot(sol.t, x, label='Koist (x)', color='blue')
plt.plot(sol.t, y, label='Dravci (y)', color='red')
plt.title('Model Lotka-Volterra: Populace dravc a koisti v ase')
plt.xlabel('as')
plt.ylabel('Populace')
plt.legend()
plt.grid(True)
plt.show()

# Fzov diagram (koist vs. dravci)
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green')
plt.title('Fzov diagram: Koist vs. Dravci')
plt.xlabel('Populace koisti (x)')
plt.ylabel('Populace dravc (y)')
plt.grid(True)
plt.show()
```
