```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('data.csv')
```

```python
goals_by_season = df['Season'].value_counts().sort_index()
goals_by_season.plot(kind='bar', color=plt.cm.tab20b.colors)
plt.title('Počet gólů podle sezóny')
plt.xlabel('Sezóna')
plt.ylabel('Počet gólů')
plt.xticks(rotation=45)
plt.show()
```

```python

competition_goals = df['Competition'].value_counts()


top_competitions = competition_goals.head(6)

plt.figure(figsize=(12, 12))
plt.pie(top_competitions, labels=top_competitions.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20c.colors)
plt.title('Top 6 soutěží podle počtu gólů')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

top_opponents = df['Opponent'].value_counts().head(15).reset_index()
top_opponents.columns = ['Team', 'Goals']

plt.figure(figsize=(12, 8))
sns.barplot(x='Goals', y='Team', data=top_opponents, hue='Team', dodge=False, palette="viridis")
plt.title('Top 15 týmů podle počtu gólů (Horizontální sloupcový graf)')
plt.xlabel('Počet gólů')
plt.ylabel('Tým')
plt.show()
```

```python

plt.figure(figsize=(14, 7))
sns.histplot(df['Minute_Clean'], bins=90, color='green', kde=True)
plt.title('Distribuce gólů v minutách zápasu')
plt.xlabel('Minuta')
plt.ylabel('Počet gólů')
plt.show()
```

```python

plt.figure(figsize=(12, 8))
scatter_plot = sns.scatterplot(x='Minute', y='Type', data=df, hue='Type', palette='bright', style='Type', s=100)
plt.title('Rozložení minut gólů podle typu střely')
plt.xlabel('Minuta')
plt.ylabel('Typ střely')
plt.legend(title='Typ střely', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```
